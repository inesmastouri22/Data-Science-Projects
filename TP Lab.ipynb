```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\house.csv")
df
```

```python
plt.plot(df['surface'], df['loyer'], 'go', markersize=4)
```

```python
regresseur = linear_model.LinearRegression()
```

```python
x = np.array(df['surface']).reshape(-1, 1)#infinite columns, and 1 line
```

```python
y = df['loyer']
```

```python
x
```

```python
y
```

```python
regresseur.fit(x,y)
```

```python
precision = regresseur.score(x,y)
```

```python
precision
```

```python
print(precision *100)
```

```python
prediction = regresseur.predict(np.array([50,70,120]).reshape(-1, 1))
```

```python
print(prediction)
```

```python
print(str(regresseur.intercept_)+'  '+str(regresseur.coef_))
```

```python
plt.plot([0, 250], 
         [regresseur.intercept_, regresseur.coef_[0] * 250 + regresseur.intercept_], 
         linestyle='--', color='blue')
```

```python
plt.show()
```
